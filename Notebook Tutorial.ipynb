### A digital lab notebook
You can compare a Jupyter notebook more or less with a lab notebook where you combine both your written (formatted) text, formulas, images and also code.

Notebooks are interactive environments where you can program and document what you are doing.

Notebooks are very useful for data science activities and writing code, because you receive direct feedback on the code you execute.

It's a backend local and/or central server to which your browser is an interactive interface, allowing the user to enter code, etc. run the code in the kernel and return the results

## 2. The different kernels
You can use different kernels (backend languages) with Jupyter - Python is one of the most used kernels, but you can easily add kernels to Jupyter. See the full list of available kernels at: https://github.com/jupyter/jupyter/wiki/Jupyter-kernels. 

### Install R and Python kernels

When using Anaconda it is very easy to use the Python and the R kernel, the 'only thing' you need to do is the following after installing Anaconda.

**create a new conda environment**

```create -n my_conda_env python=3.6```

**activate the conda environment**

```activate my_conda_env```

or 

```source activate my_conda_env```


**install jupyter**

```conda install jupyter```

**install r-essentials**

```conda install r-essentials```

**start jupyter**

```jupyter notebook```
