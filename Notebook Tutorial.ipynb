## 3. Basic navigation through a notebook
It's extremely useful to get yourself acquanted to the most import keyboard shortcuts used in Jupyter notebooks. 

The Jupyter Notebook has two different keyboard input modes. **Edit mode** allows you to type code or text into a cell and is indicated by a green cell border. **Command mode** binds the keyboard to notebook level commands and is indicated by a grey cell border with a blue left margin.

### Switching between edit and command mode
Switching between edit and command mode can be done in different ways:
* command --> edit: 
  * double click the cell
  * press enter
* edit --> command:
  * ESC : no execution / rendering
  * CTRL + ENTER or SHIFT + ENTER: execution of code cell, rendering of markdown

### Basic movement
To move the active cell (the one you are editing in) up or down you can use the up/down arrow keys on the keyboard. However if you are familiar with vim movement keys (you better get used to them, cause they are sooooo handy and efficient - once you remember them) you can also use j/k.

## 11. Widgets

````python
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
````

````python
t = np.arange(0.0, 1.0, 0.01)

%matplotlib inline

def pltsin(f,c):
    plt.plot(np.sin(2*np.pi*t*f), color=c)
    plt.show();
    
interact(pltsin, f=widgets.FloatSlider(
    value=7.5,
    min=0,
    max=10.0,
    step=0.1,
    description='Test:',
    disabled=False,
    continuous_update=False,
    orientation='vertical',
    readout=True,
    readout_format='.1f',
), c=widgets.Dropdown(
    options={'Blue': 'blue', 'Red': 'red', 'Green': 'green'},
    value='red',
    description='Color:',
));
````
